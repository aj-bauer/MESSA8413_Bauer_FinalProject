## PART 0: Download data from API

```python
# Import data: First 300 K rows
# Dataset documentation: https://data.cityofnewyork.us/Social-Services/311-Service-Requests-from-2010-to-Present/erm2-nwe9/about_data

# NOTE: THIS CODE MAY TAKE MORE THAN AN HOUR TO RUN IN ITS ENTIRETY

print('Beginning process...')
time_start = time.time()

# Define parameters
url = 'https://data.cityofnewyork.us/api/v3/views/erm2-nwe9/query.json'
header = {'X-App-Token': 'S7ElOhAWdJM60qQzSVzSJP8AR'}
ranges = [range(10),
          range(10,20),
          range(20,30),
          range(30,40),
          range(40,50),
          range(50,60),
          range(60,70),
          range(70,80),
          range(80,90),
          range(90,100)]

# Iterate through 15 files
for i in range(9,10):

  print(f'Beggining file {str(i)} download...')
  time0 = time.time()

  # Define request
  json_q = [{"query": "SELECT *",
          "page": {"pageNumber":n+1,
                  "pageSize": 3000},
          "includeSynthetic": False} for n in ranges[i]]

  # Download files as JSON
  x = []
  for j in json_q:
    time1 = time.time()
    x.append(requests.post(url, headers=header, json = j).text)
    print(f'\tPage {j['page']['pageNumber']} downloaded ({(time.time() - time1):.4f} sec)...')

  print(f'\tAll pages downloaded ({(time.time() - time0):.4f} sec)...')

  # Convert to dataframe(s)
  for k in range(len(x)):
    x[k] = pd.DataFrame(json.loads(x[k]))

  # Concatenate dataframes
  df = pd.concat(x)

  # export data to CSV
  df.to_csv(f'..\\Data\\0_NYC311_raw\\NYC311_{str(i)}.csv', index=False)
  print(f'File {str(i)} exported ({(time.time() - time0):.4f} sec)\n')

time_finish = time.time()
print(f'Downloads complete: {(time_finish - time_start):.4f} seconds.')
```

## PART 1(B): SQL DDL

```python
# NOTE: Prior to executing this section, create the database using the following SQL:
initialization = '''
    DROP DATABASE IF EXISTS mesa8413;
    
    CREATE DATABASE mesa8413
        WITH
        OWNER = postgres
        ENCODING = 'UTF8'
        LC_COLLATE = 'English_United States.1252'
        LC_CTYPE = 'English_United States.1252'
        LOCALE_PROVIDER = 'libc'
        TABLESPACE = pg_default
        CONNECTION LIMIT = -1
        IS_TEMPLATE = False;
    '''


# Establish Connection
time0 = time.time()
with psycopg.connect(dbname="mesa8413", user="postgres", password="MESA8413") as conn:
    print(f'DATABASE CONNECTED: {time.time() - time0} SEC\n')

    # Define DDL statements
    pre_clear = 'DROP SCHEMA IF EXISTS nyc311 CASCADE'
    ddl = '''
    -- Create Schema
    CREATE SCHEMA nyc311
    	
    	-- Define tables
    	CREATE TABLE agency (
    		agency varchar PRIMARY KEY,
    		agency_name varchar NOT NULL
    	)
    	CREATE TABLE call_center (
    		cc_id varchar PRIMARY KEY,
    		cc_street varchar NOT NULL,
    		cc_borough varchar NOT NULL
    	)
    	CREATE TABLE call_operator (
    		op_id varchar PRIMARY KEY,
    		op_name varchar NOT NULL,
    		op_ssn character(9) NOT NULL,
    		op_salary numeric(7,2) NOT NULL,
    		op_cc varchar NOT NULL REFERENCES call_center (cc_id) 
    	)
    	CREATE TABLE complaint_type (
    		complaint_id varchar PRIMARY KEY,
    		agency varchar NOT NULL REFERENCES agency (agency),
    		complaint_type varchar NOT NULL,
    		complaint_descriptor varchar,
    		location_type varchar
    	)
    	CREATE TABLE complaint_incident (
    		unique_key character(8) PRIMARY KEY,
    		complaint_id varchar NOT NULL REFERENCES complaint_type (complaint_id),
    		op_id varchar NOT NULL REFERENCES call_operator (op_id),
    		created_date timestamp NOT NULL,
    		incident_zip integer,
    		incident_address varchar,
    		cross_street_1 varchar,
    		cross_street_2 varchar,
    		intersection_street_1 varchar,
    		intersection_street_2 varchar,
    		address_type varchar,
    		city varchar,
    		landmark varchar,
    		community_board varchar NOT NULL,
    		borough varchar NOT NULL,
    		x_coordinate_state_plane integer,
    		y_coordinate_state_plane integer,
    		open_data_channel_type varchar NOT NULL,
    		park_facility_name varchar,
    		latitude numeric(8,6),
    		longitude numeric(8,6)
    	)
    	CREATE TABLE complaint_status (
    		status_id character(15) PRIMARY KEY,
    		unique_key character(8) NOT NULL REFERENCES complaint_incident (unique_key) ON DELETE CASCADE,
    		status varchar NOT NULL,
    		resolution_description varchar,
    		resolution_action_updated_date timestamp,
    		due_date timestamp,
    		closed_date timestamp
    	)
    	CREATE TABLE taxi_details (
    		taxi_id character(13) PRIMARY KEY,
    		unique_key character(8) NOT NULL REFERENCES complaint_incident (unique_key) ON DELETE CASCADE,
    		taxi_company_borough varchar,
    		taxi_pick_up_location varchar
    	)
    	CREATE TABLE road_details (
    		road_id character(13) PRIMARY KEY,
    		unique_key character(8) NOT NULL REFERENCES complaint_incident (unique_key) ON DELETE CASCADE,
    		bridge_highway_name varchar NOT NULL,
    		bridge_highway_direction varchar,
    		bridge_highway_segment varchar,
    		road_ramp varchar
    	)
    
    	-- Define indexes
    	CREATE INDEX incident_date_idx ON complaint_incident (created_date)
    	CREATE INDEX incident_board_idx ON complaint_incident (community_board)
    	CREATE INDEX op_ssn_idx ON call_operator (op_ssn);
        '''

    # Establish cursor
    with conn.cursor() as cur:

        # Execute DDL Statements
        cur.execute(pre_clear)
        cur.execute(ddl)

        # Commit changes
        conn.commit()
print(f'SCHEMA CREATED: {time.time() - time0} SEC')
```

### Index Justification:
- By creating an index on complaint_incident.created date, the user can quickly access time ranges (e.g. all data since Nov. 10, or,
  as we will explore later, all complaints between 3 and 4pm on October 30). Without an ordered list of timestamps, a query execution
  would rely on a full scan to identify complaints within the desired time range, so indexing can improve performance.
- By creating a clustered index on complaint_incident.community_board, we can quickly access all complaints for a given region/neighborhood
  of the city, avoiding a full scan in that regard.
- The hypothetical HR department that hires & manages the call center employees may need to retrieve operators by social security number.
  Thus adding an index allows for a query to be performed via binary search instead of a full scan.
